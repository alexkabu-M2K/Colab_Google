```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Dados de exemplo
dados = {

         'Altura': [170, 180, 175, 160, 165],
         'Peso': [65, 80, 75, 60, 65],
 }

# Criar um DataFrame a partir dos dados
df = pd.DataFrame(dados)

# Aplicando KDE
sns.kdeplot(x=df['Altura'], y=df['Peso'], cmap="Reds", bw_adjust=0.5)
```
